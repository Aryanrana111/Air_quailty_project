Which month in each year recorded the highest average PM2.5 levels?

Is there a correlation between PM10 and AQI?

```python
cor = df[["PM10","AQI"]]
cor
```

```python
import matplotlib.pyplot as plt
```

```python
correlation = cor["AQI"].corr(cor["PM10"])
print(f"Correlation between PM10 and AQI,(Correlation*100)%)")
plt.scatter(cor["PM10"],cor["AQI"])
plt.title("AQI vs PM10")
plt.xlabel("PM10")
plt.ylabel("AQI")
plt.show()
```

How does AQI vary on holidays compared to non-holidays?

```python
holiday = df[df["Holidays_Count"] == 1]
non_holiday = df[df["Holidays_Count"] == 0]
```

```python
avg1 = holiday["AQI"].mean()
avg2 = non_holiday["AQI"].mean()
```

```python
print("Holiday AQI Average")
print(avg1)
print()
print("non-Holiday AQI Average")
print(avg2)
```

```python
li = [avg1 , avg2]
label = ["Holidays","Non-Holidays1"]
```

```python
plt.pie(li , labels = label , autopct = "%.2f" , radius = 0.7 ,shadow = True)
plt.show()
```

Which pollutant shows the strongest correlation with AQI

```python
cor = df[["PM2.5","PM10","NO2","SO2","CO","Ozone","AQI"]]
cor
```

```python
correlation = cor["AQI"].corr(cor["PM10"])
print(f"correlation B/W PM2 AQI is {correlation*100}%")

correlation = cor["AQI"].corr(cor["PM2.5"])
print(f"correlation B/W PM2.5 AQI is {correlation*100}%")

correlation = cor["AQI"].corr(cor["NO2"])
print(f"correlation B/W NO2 AQI is {correlation*100}%")

correlation = cor["AQI"].corr(cor["SO2"])
print(f"correlation B/W SO2 AQI is {correlation*100}%")

correlation = cor["AQI"].corr(cor["CO"])
print(f"correlation B/W CO AQI is {correlation*100}%")

correlation = cor["AQI"].corr(cor["Ozone"])
print(f"correlation B/W Ozone AQI is {correlation*100}%")
print()


fig,ax=plt.subplots(3,2)
ax[0,0].scatter(cor["PM10"],cor["AQI"],marker=".")
ax[0,0].set_title("AQI vs PM10")
ax[0,0].set_xlabel("PM10")
ax[0,0].set_ylabel("AQI")

ax[0,1].scatter(cor["PM2.5"],cor["AQI"],marker=".")
ax[0,1].set_title("AQI vs PM2.5")
ax[0,1].set_xlabel("PM2.5")
ax[0,1].set_ylabel("AQI")

ax[1,0].scatter(cor["NO2"],cor["AQI"],marker=".")
ax[1,0].set_title("AQI vs NO2")
ax[1,0].set_xlabel("NO2")
ax[1,0].set_ylabel("AQI")

ax[1,1].scatter(cor["SO2"],cor["AQI"],marker=".")
ax[1,1].set_title("AQI vs SO2")
ax[1,1].set_xlabel("SO2")
ax[1,1].set_ylabel("AQI")

ax[2,0].scatter(cor["CO"],cor["AQI"],marker=".")
ax[2,0].set_title("AQI vs CO")
ax[2,0].set_xlabel("CO")
ax[2,0].set_ylabel("AQI")

ax[2,1].scatter(cor["Ozone"],cor["AQI"],marker = ".")
ax[2,1].set_title("AQI vs Ozone")
ax[2,1].set_xlabel("Ozone")
ax[2,1].set_ylabel("AQI")

plt.tight_layout()
```
